```python
import re
from collections import Counter
import numpy as np
```

```python
def tokenize(text):
    text = text.lower()
    text = re.sub(r"<.*?>", "", text)
    text = re.sub(r"[^a-zA-Z]", " ", text)
    return text.split()
```

```python
tokenize("the cat sat on the mat")
```

```python
tokenize("the cat sat on the mat !!!!")
```

```python
def build_vocab(texts, max_words=100):
    counter = Counter()

    for text in texts:
        counter.update(tokenize(text))

    most_common = counter.most_common(max_words)

    vocab = {"<pad>": 0, "<unk>": 1}
    for i, (word, _) in enumerate(most_common, start=2):
        vocab[word] = i

    return vocab
```

```python
vocab = build_vocab(["the cat sat on the mat", "the cat is fat"])
vocab
```

```python
def encode(vocab, text, max_len=100):
    tokens = tokenize(text)
    ids = [vocab.get(token, vocab["<unk>"]) for token in tokens]
    return ids[:max_len]
```

```python
encode(vocab, "the mat is a cat")
```

```python
def one_hot_encode(vocab, token):
    vector = np.zeros((len(vocab),)) 
    token_index = vocab[token] 
    vector[token_index] = 1
    return vector 
```

```python
one_hot_encode(vocab, "mat")
```

```python
def build_inverse_dict (vocab):
    return dict( (v, k) for k, v in vocab.items())
```

```python
def decode(vocab, int_sequence, inv_vocab = None):
    if inv_vocab is None: 
        inverse_vocabulary = build_inverse_dict(vocab)
    else: inverse_vocabulary = inv_vocab
    return " ".join(
        inverse_vocabulary.get(i, "[UNK]") for i in int_sequence)
```

```python
decode(vocab, [2,6,7])
```
