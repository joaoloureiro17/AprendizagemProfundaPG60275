```python
## Example of XNOR with PyTorch
```

```python
## !pip install torch
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
## Dataset: XNOR truth table
X = torch.tensor([
    [0., 0.],
    [0., 1.],
    [1., 0.],
    [1., 1.]
])
y = torch.tensor([
    [1.],
    [0.],
    [0.],
    [1.]
])
```

```python
# Neural network model
class XNORNet(nn.Module):
    def __init__(self, hidden = 2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(2, hidden),   # hidden layer - 4 nodes
            nn.Sigmoid(),      # nonlinearity (important!)
            nn.Linear(hidden, 1),   # output layer - 1 node
            nn.Sigmoid()       # output as probability
        )

    def forward(self, x):
        return self.net(x)

model = XNORNet(4)
```

```python
print(model)
```

```python
# Loss and optimizer
criterion = nn.BCELoss() # binary cross entropy
optimizer = optim.Adam(model.parameters(), lr=0.05) # training algorithm
```

```python
# Training loop
epochs = 100
for epoch in range(epochs):
    # forward
    y_pred = model(X)
    loss = criterion(y_pred, y)

    # backward
    optimizer.zero_grad()
    loss.backward() # calcular as derivadas
    optimizer.step() # coisar os pesos

    # print progress
    if epoch % 20 == 0:
        print(f"Epoch {epoch:4d} | Loss: {loss.item():.4f}")
```

```python
# Test the trained model
with torch.no_grad(): # estamos a prever sem guardar nenhuma info, estamos so a calcular a saida
    preds = model(X)
    print("\nPredictions:")
    print(preds.round())
```
